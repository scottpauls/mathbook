```python
def iterative_topic_extraction(topic_model, docs, min_items=15):
    """
    Iteratively extracts topics using BERTopic.

    Args:
        topic_model: The BERTopic model instance.
        docs: The list of documents.
        min_items: The minimum number of items in a topic to trigger further extraction.

    Returns:
        A dictionary mapping topic IDs to a list of subtopics or documents.
    """

    topic_info = topic_model.get_topic_info()
    topic_results = {}  # Store results for each topic

    for index, row in topic_info.iterrows():
        topic_id = row['Topic']
        if topic_id != -1 and row['Count'] > min_items:  # Exclude outliers and small topics
            # Get documents belonging to this topic
            topic_docs_indices = [i for i, t in enumerate(topics) if t == topic_id]
            topic_docs = [docs[i] for i in topic_docs_indices]

            # Apply BERTopic again to the subset of documents
            sub_topic_model = BERTopic()
            sub_topics, sub_probs = sub_topic_model.fit_transform(topic_docs)

            # Store subtopics or original documents in the results
            topic_results[topic_id] = {
                'sub_topics': sub_topic_model.get_topic_info(),
                'documents': topic_docs
            }
        else:
            # Store original documents for topics not meeting the criteria
            topic_docs_indices = [i for i, t in enumerate(topics) if t == topic_id]
            topic_docs = [docs[i] for i in topic_docs_indices]
            topic_results[topic_id] = {'documents': topic_docs}

    return topic_results

# Apply the function
topic_results = iterative_topic_extraction(topic_model, docs)

# Access results for each topic
for topic_id, results in topic_results.items():
    print(f"Topic {topic_id}:")
    if 'sub_topics' in results:
        print("  Subtopics:", results['sub_topics'])
    print("  Documents:", results['documents'][:5])  # Print first 5 documents for brevity
```
